```python
import redis
import json
```

```python
r = redis.StrictRedis(host='localhost', port=6379, db=0)
```

```python
# I selected the same weather DB that was created in previous task
# extract data from json file and load it to Redis DB
with open('weather.json',"r") as file:
    data = json.load(file)
```

```python
for entry in data:
    key = entry["Dates"]
    value = json.dumps(entry)
    r.set(key, value)
```

```python
# display data from Redis DB:
keys = r.keys("*")  # Get all keys in the database
for key in keys:
    value = r.get(key)
    if value:
        data = json.loads(value)
        print("Key:", key.decode("utf-8"))
        print("Value:", data)
        print("----------------------")
```

```python
# Set a key with a specific value in Redis.
value = {
    "Dates": "2023-01-03",
    "Temperature mean": 5.6,
    "Temperature max": 9.2,
    "Temperature min": 1.7,
    "Rain": 2.4,
    "Snowfall": 0.0,
    "Wind direction": 269.0
}

# Convert data to JSON format
json_data = json.dumps(data)

key = "2023-07-15"
r.set(key, json_data)
```

```python
# Retrieve the value of a key from Redis.
value = r.get("2023-07-15")
```

```python
#Delete a key-value pair from Redis.
key_to_delete = "2023-06-15"
deleted_count = r.delete(key_to_delete)

if deleted_count == 1:
    print("Key", key_to_delete, "deleted successfully.")
else:
    print("Key", key_to_delete, "not found.")
```

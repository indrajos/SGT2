```python
import requests
import json
import pandas as pd
```

```python
url_temp = 'https://archive-api.open-meteo.com/v1/archive?latitude=54.69&longitude=25.28&start_date=2023-01-01&end_date=2023-06-18&daily=temperature_2m_max,temperature_2m_min,temperature_2m_mean&timezone=Europe%2FBerlin&min=2023-06-04&max=2023-06-18'
response = requests.get(url_temp)
temperature = response.json()
```

```python
temperature
```

```python
temp_dates = list(temperature['daily']['time'])
temperature_max = list(temperature['daily']['temperature_2m_max'])
temperature_min = list(temperature['daily']['temperature_2m_min'])
temperature_mean = list(temperature['daily']['temperature_2m_mean'])
```

```python
temp_data = {"Dates":temp_dates,"Temperature mean":temperature_mean,"Temperature max":temperature_max,"Temperature min":temperature_min}
```

```python
temp_df = pd.DataFrame(temp_data)
temp_df.head()
```

```python
url_fall = 'https://archive-api.open-meteo.com/v1/archive?latitude=54.69&longitude=25.28&start_date=2023-01-01&end_date=2023-06-18&daily=rain_sum,snowfall_sum,winddirection_10m_dominant&timezone=Europe%2FBerlin&min=2023-06-04&max=2023-06-18'
response = requests.get(url_fall)
fall = response.json()
```

```python
fall
```

```python
fall_dates = list(fall['daily']['time'])
rain = list(fall['daily']['rain_sum'])
snowfall = list(fall['daily']['snowfall_sum'])
wind = list(fall['daily']['winddirection_10m_dominant'])
```

```python
fall_data = {"Dates":fall_dates,"Rain":rain,"Snowfall":snowfall,"Wind direction":wind}
```

```python
fall_df = pd.DataFrame(fall_data)
fall_df.head()
```

```python
from sqlalchemy import create_engine
from sqlalchemy import text
import mysql.connector
```

```python
cnx = mysql.connector.connect(user='root',
                              password='xxx',
                              host='localhost',
                              database='weather')

cursor = cnx.cursor()
engine = create_engine('mysql+mysqlconnector://root:%s@localhost:3306/weather' % quote('xxx'))
temp_df.to_sql('temperature', engine, index=False, if_exists="replace")
fall_df.to_sql('fall', engine, index=False, if_exists="replace")

cnx.commit()

cursor.close()
cnx.close()
```

```python
with engine.begin() as con:
    temp = pd.read_sql(text("SELECT * FROM temperature"),con)
```

```python
with engine.begin() as con:
    rain = pd.read_sql(text("SELECT * FROM fall"),con)
```
